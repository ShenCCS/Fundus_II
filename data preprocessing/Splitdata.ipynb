```python
#1. move together

File = 'validation'

import os
import shutil

# Training validation Testing
# source_folder = '/home/john/network/RETFound/data/dataset/Testing'
# destination_folder = '/home/john/network/RETFound/data/Image_collection/Testing'

source_folder = '/home/john/network/RETFound/data/dataset/' + File
destination_folder = '/home/john/network/RETFound/data/Image_collection/' + File


def move_file(source_folder,destination_folder):
    count = 0
    for root,sub,files in os.walk(source_folder):
        for file in files:
            source_folder_path = os.path.join(root,file)
            destination_file_path = os.path.join(destination_folder,file)
            shutil.copy2(source_folder_path,destination_file_path)
            count += 1
    print("Image files number: ",count)

move_file(source_folder,destination_folder)
print("Done")
```
